## 1.4 Cleaning our Dataset

we aregoing to perform data cleaning procedures.

### 1.41) Validity


```
# Procedure 1: dropping the un named column
# Data Cleaning Action: checking for null values across all rows
# Explanation: we dont have null values
#
forex.drop(['Unnamed: 0'], axis = 1, inplace=True)
```

```
# Procedure 1:
# Data Cleaning Action: checking for null values across all rows
# Explanation: we dont have null values
#
forex.isnull().sum()
```

```
# Procedure 2: replacing ND
# Data Cleaning Action: replaceing ND with zeros
# Explanation:
forex_nd =forex.replace(to_replace = 'ND',value = None)

forex_nd = forex_nd.dropna(how= "all")
forex_nd.head(10)
```

```
forex_nd.head(13)
```

```
forex_nd.isnull().any()
```

### 1.44) Consitency: <font color="green">Challenges</font>

```
forex_nd.info()
```

### 1.45) Uniformity:


```
#Procedure 1: changing the column names
# 
# 
forex_nd.columns =['Time_Serie','Australia','Euro','New_zealand','United_kingdom','Brazil','Canada','China','Homg_kong','India','Korea','Mexico','South_africa','Singapore','Denmark','Japan','Malaysia','Norway','Sweden','Sri_lanka','Switzerland','Taiwan','Thailand']
forex_nd.head()
```

```
forex_nd['Australia'] = forex_nd.Australia.str.rstrip('').astype(float)
forex_nd['Euro'] = forex_nd.Euro.str.strip('').astype(float)
forex_nd['New_zealand'] = forex_nd.New_zealand.str.strip('').astype(float)
forex_nd['United_kingdom'] = forex_nd.United_kingdom.str.strip('').astype(float)
forex_nd['Brazil'] = forex_nd.Brazil.str.strip('').astype(float)
forex_nd['Canada'] = forex_nd.Canada.str.strip('').astype(float)
forex_nd['China'] = forex_nd.China.str.strip('').astype(float)
forex_nd['Homg_kong'] = forex_nd.Homg_kong.str.strip('').astype(float)
forex_nd['India'] = forex_nd.India.str.strip('').astype(float)
forex_nd['Korea'] = forex_nd.Korea.str.strip('').astype(float)
forex_nd['Mexico'] = forex_nd.Mexico.str.strip('').astype(float)
forex_nd['South_Africa'] = forex_nd.South_africa.str.strip('').astype(float)
forex_nd['Singapore'] = forex_nd.Singapore.str.strip('').astype(float)
forex_nd['Denmark'] = forex_nd.Denmark.str.strip('').astype(float)
forex_nd['Japan'] = forex_nd.Japan.str.strip('').astype(float)
forex_nd['Malaysia'] = forex_nd.Malaysia.str.strip('').astype(float)
forex_nd['Norway'] = forex_nd.Norway.str.strip('').astype(float)
forex_nd['Sweden'] = forex_nd.Sweden.str.strip('').astype(float)
forex_nd['Sri_lanka'] = forex_nd.Sri_lanka.str.strip('').astype(float)
forex_nd['Switzerland'] = forex_nd.Switzerland.str.strip('').astype(float)
forex_nd['Taiwan'] = forex_nd.Taiwan.str.strip('').astype(float)
```

```
forex_nd.drop_duplicated()
```
